```python
import pandas as pd
import numpy as np
```

```python
# Load the Dataset: Import the dataset using pandas. 
df=pd.read_csv('C:/Users/abhay.manhas/Downloads/movies_sample.csv')
```

```python
# Steps for Data Cleaning 
# Load the Dataset: Import the dataset using pandas. 
# Inspect the Data: Check the first few rows and summary statistics. 
# Handle Null Values: Fill or drop null values. 
# Correct Data Types: Convert columns to appropriate data types. 

# Handle Negative Values: Replace or remove negative values. 
# Remove Duplicates: Ensure no duplicate rows are present. 
# Standardize Data: Ensure consistency in categorical data.
```

```python
# Handle Null Values: Fill or drop null values.
df = df.dropna(subset=['Genre'])
df = df.dropna(subset=['Director'])
df = df.dropna(subset=['Title'])
df[['Revenue (Millions)', 'Budget (Millions)', 'Rating', 'Year']] = df[['Revenue (Millions)', 'Budget (Millions)', 'Rating', 'Year']].fillna(0)

print(df.isnull().sum())
```

```python
# Correct Data Types: Convert columns to appropriate data types. 
df['Genre']=df['Genre'].astype(str)
df[['Revenue (Millions)', 'Budget (Millions)','Rating']] = df[['Revenue (Millions)', 'Budget (Millions)','Rating']].astype(float)
```

```python
# as year was not able to convert so 
df['Year'].str.replace('Unknown','0')
```

```python
# Remove Duplicates: Ensure no duplicate rows are present. 
df.drop_duplicates()
```

```python
# Handle Negative Values: Replace or remove negative values. 

df['Budget (Millions)'] = df['Budget (Millions)'].apply(lambda x: 0 if x < 0 else x)  # Replace negative values in Budget column with 0
df['Revenue (Millions)'] = df['Revenue (Millions)'].apply(lambda x: 0 if x < 0 else x)
print(df.min()) # Check for negative values
```

```python
# Inspect the Data: Check the first few rows and summary statistics. 
df.groupby(['Genre']).agg({'Rating':['mean','min','max','count'],'Revenue (Millions)':['mean','min','max','count'],'Budget (Millions)':['mean','min','max','count']})
```

```python
df.to_csv('movie.csv', index=False)
```
