```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('C:/Users/abhay.manhas/Downloads/banking_sample.csv')
```

```python
df.drop_duplicates()
```

```python
# Handling Missing Values: 
# Age: Replaced missing values with the mean age. 
# Gender: Filled missing values with the most frequent value (mode). 
# AccountBalance: Replaced missing values with 0. 
# AccountType: Replaced missing values with 'Unknown'. 
# AccountOpenDate: Filled missing dates with a placeholder date. 
# Email: Filled missing emails with a placeholder email. 
# Correcting Bad Data: 
# AccountType: Replaced invalid account types with 'Unknown'. 
# Email: Validated and corrected email addresses using a regular expression.

```

```python
# Age: Replaced missing values with the mean age. 
avgage=df['Age'].mean()
df['Age']=df['Age'].fillna(avgage)
```

```python
# to check value common use and then replace
# Gender: Filled missing values with the most frequent value (mode). 
df['Gender'].mode()
df['Gender']=df['Gender'].fillna(0)
df['Gender']=df['Gender'].replace(0,'Male')
```

```python
# AccountBalance: Replaced missing values with 0. 
df['AccountBalance']=df['AccountBalance'].fillna(0)
```

```python
# AccountType: Replaced missing values with 'Unknown'. 
df['AccountType']=df['AccountType'].fillna(0)
df['AccountType']=df['AccountType'].replace(0,'Unknown')
```

```python
df
```

```python
# Fill missing AccountOpenDate with a placeholder date
df['AccountOpenDate'].fillna('0000-00-00', inplace=True)

# Fill missing Email with a placeholder email
df['Email'].fillna('unknown@email.com', inplace=True)
```

```python
df
```

```python
# Correcting Bad Data: 

df['Name'].fillna('delete', inplace=True)
```

```python
# AccountType: Replaced invalid account types with 'Unknown'. 
df['AccountType']=df['AccountType'].str.replace('Invalid','Unknown')
```

```python

# Email: Validated and corrected email addresses using a regular expression.
import re

mailregex = '^[a-zA-Z0-9.]+@[a-zA-Z]+\.[a-zA-Z]+$'

def validate_email(email):
    if re.match(mailregex, email):
        return email
    else:
    
        return email + '.com' if not email.endswith('.com') else 'invalid_email@example.com'

df['Email'] = df['Email'].apply(validate_email)
```

```python
# Correcting Bad Data: 
df.drop(df[df['Name'] == 'delete'].index, inplace=True)
```

```python
# to assign null
df['CustomerID'].fillna('NA',inplace=True)
```

```python
df.to_csv('Banking.csv', index=False)
```
